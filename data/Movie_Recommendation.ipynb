```python
import ast

def convert(text):
    l = []
    if isinstance(text, str):
        try:
            for i in ast.literal_eval(text):
                l.append(i['name'])
        except (ValueError, SyntaxError):
            # If it's a malformed string or not a literal, we'll just pass
            # This case shouldn't be hit with the current data format after the first run
            pass
    elif isinstance(text, list): # Handles cases where the column is already a list
        for i in text:
            if isinstance(i, dict) and 'name' in i:
                l.append(i['name'])
            elif isinstance(i, str):
                l.append(i)
    return l
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
import ast

def convert_cast(text):
    l = []
    counter = 0
    if isinstance(text, str):
        try:
            for i in ast.literal_eval(text):
                if counter < 3:
                    l.append(i['name'])
                counter += 1
        except (ValueError, SyntaxError):
            pass
    elif isinstance(text, list): # Handles cases where the column is already a list
        for i in text:
            if counter < 3:
                if isinstance(i, dict) and 'name' in i:
                    l.append(i['name'])
                elif isinstance(i, str):
                    l.append(i)
            counter += 1
    return l

def fetch_director(text):
    l = []
    if isinstance(text, str):
        try:
            # Use json.loads for more robust parsing of JSON strings
            for i in json.loads(text):
                if i['job'] == 'Director':
                    l.append(i['name'])
                    break # Assuming only one director
        except (json.JSONDecodeError, ValueError):
            # Catch JSON decoding errors as well as general ValueErrors
            pass
    elif isinstance(text, list):
        for i in text:
            if isinstance(i, dict) and i.get('job') == 'Director':
                l.append(i['name'])
                break
    return l
```

```python
movies['cast'] = movies['cast'].apply(convert_cast)
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
def process_tags(tag_list):
    processed_words = []
    # First element is the overview sentence
    overview_sentence = tag_list[0]
    # Split overview into words and add to processed_words
    processed_words.extend([word.lower() for word in overview_sentence.split()])

    # Process remaining elements (genres, keywords, cast, crew)
    for item in tag_list[1:]:
        cleaned_item = item.lower().replace(" ", "")
        processed_words.append(cleaned_item)
    return ", ".join(processed_words)

movies['tags'] = movies['tags'].apply(process_tags)
```

```python
%pip install nltk

import nltk
from nltk.stem import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
def stems(text):
    T = []

    for i in text.split():
        T.append(ps.stem(i))

    return " ".join(T)
```

```python

new_df['tags'] = new_df['tags'].apply(stems)
```

```python
new_df.iloc[0]['tags']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(new_df['tags']).toarray()
```

```python
vector[0]
```

```python
vector.shape
```

```python
len(cv.get_feature_names_out())
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity.shape
```

```python
new_df[new_df['title'] == 'The Lego Movie'].index[0]
```

```python
def recommend(movie):
    index = new_df[new_df['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Spider-Man 2')
```
